```python
start= time.time()
def binatodeci(binary):
    return sum(val*(2**idx) for idx, val in enumerate(reversed(binary)))

print(binatodeci(a))
print(binatodeci(b))
print(binatodeci(a)+binatodeci(b))

end = time.time()
print("Duration: ",end-start," Seconds")
```

```python
def addbits(t,v,ca):
    if t + v + ca == 3:
        ca,out = 1,1
    elif t + v + ca == 2:
        ca,out = 1,0
    elif t + v + ca == 1:
        ca,out = 0,1
    else:
        ca,out = 0,0

    return [int(out),int(ca)]

def addbins(a,b):
    x = a.copy()
    y = b.copy()
    x.reverse()
    y.reverse()
    if len(a)>len(b):
        smaller = len(b)
        bigger = len(a)
        zz = x.copy()
    else:
        smaller = len(a)
        bigger = len(b)
        zz = y.copy()
    print("Bigger:",bigger)
    print("Smaller:", smaller)
    z = [None] * bigger
    carry = 0
    for i in range(smaller):
        #print(i)
        xx = x[i]
        yy = y[i]
        z[i], carry = addbits(x[i],y[i],carry)
    if bigger-smaller > 0:
        for j in range(smaller,bigger):
            z[j],carry  = addbits(zz[j],0,carry)
    #print(z)
    z.reverse()
    #print(z)
    return z

```

```python
start = time.time()
binaryaddition = addbins(a,b)
print(a)
print("_----------------------------",b)
print("Equals:")

print(binaryaddition)
end = time.time()
print("Duration: ",end-start," Seconds")
```

```python

# function to subtract two values
# using 2's complement method
def Subtract(a,b):

    # ~b is the 1's Complement of b
    # adding 1 to it make it 2's Complement
    c = a + (~b + 1)
    return c

# Driver code
if __name__ == "__main__" :

    # multiple assignments
    a,b = 2,3
    print(Subtract(a,b))

    a,b = 9,7
    print(Subtract(a,b))
```
